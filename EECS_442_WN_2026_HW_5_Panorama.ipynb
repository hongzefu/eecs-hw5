# Problem 1: Homography Estimation

First, we will implement a function that computes a homography matrix $H$ from a set of corresponding points between two images. We'll estimate $H$ using two methods: 1) an analytic (linear) method and 2) a nonlinear least squares optimization method. For a pair of corresponding points, $p_i$ and $p_i'$, we solve for a homography matrix $H$ such that $p_i' = H p_i$:

$$
\left[\begin{array}{c}{x'} \\ {y'} \\ {1}\end{array}\right] = \left[\begin{array}{lll}{a} & {b} & {c} \\ {d} & {e} & {f} \\ {g} & {h} & {i}\end{array}\right] \cdot\left[\begin{array}{l}{x} \\ {y} \\ {1}\end{array}\right]
$$


We will work with two sets of 2D points representing the shape of the University of Michigan’s "M" logo, but at different orientations (see below for visualization). One set serves as the original points, while the other represents the destination points. Our goal is to compute the homography that maps the original points to the destination points, enabling us to warp the original shape accordingly.

### (a) For a pair of corresponding points $p_i$ and $p_i'$, obtain an analytic solution for $H$.


Let

$$
p_i=\begin{bmatrix}x_i\\y_i\\1\end{bmatrix}, \qquad
p_i'=\begin{bmatrix}x_i'\\y_i'\\1\end{bmatrix}, \qquad
H=
\begin{bmatrix}
a & b & c\\
d & e & f\\
g & h & i
\end{bmatrix}.
$$

Since homography is defined in homogeneous coordinates,

$$
p_i' \sim H p_i.
$$

Thus,

$$
x_i'=\frac{ax_i+by_i+c}{gx_i+hy_i+i}, \qquad
y_i'=\frac{dx_i+ey_i+f}{gx_i+hy_i+i}.
$$

Rearranging gives two linear equations:

$$
ax_i+by_i+c-x_i'(gx_i+hy_i+i)=0,
$$

$$
dx_i+ey_i+f-y_i'(gx_i+hy_i+i)=0.
$$

This can be written as

$$
\begin{bmatrix}
x_i & y_i & 1 & 0 & 0 & 0 & -x_i x_i' & -y_i x_i' & -x_i' \\
0 & 0 & 0 & x_i & y_i & 1 & -x_i y_i' & -y_i y_i' & -y_i'
\end{bmatrix}
\mathbf{h}=0,
$$

where

$$
\mathbf{h}=
\begin{bmatrix}
a & b & c & d & e & f & g & h & i
\end{bmatrix}^T.
$$

Stacking all point correspondences gives

$$
A\mathbf{h}=0.
$$

The analytic solution is obtained by SVD: $\mathbf{h}$ is the right singular vector of $A$ corresponding to the smallest singular value. Reshaping $\mathbf{h}$ gives the homography matrix $H$, then we normalize it so that $i=1$.

At least 4 corresponding point pairs are needed, since one pair gives only 2 equations while $H$ has 8 degrees of freedom up to scale.

```python
def fit_homography(pts1 ,pts2):
    """
    Given a set of N correspondences of the form [x,y], [x',y'],
    fit a homography that maps [x,y,1] to [x',y',1] using the analytic solution.

    Input:
        pts1, pts2: (N, 2) matrices representing N corresponding points [x, y] and [x', y']
    Returns:
        H: a (3,3) homography matrix that satisfies [x',y',1] == H [x,y,1]
    """

    H = None

    #############################################################################
    #                                   TODO                                    #
    #############################################################################
    if pts1.shape != pts2.shape:
        raise ValueError('pts1 and pts2 must have the same shape')
    if pts1.shape[0] < 4:
        raise ValueError('At least 4 point correspondences are required')

    A = []
    for (x, y), (xp, yp) in zip(pts1, pts2):
        A.append([x, y, 1, 0, 0, 0, -x * xp, -y * xp, -xp])
        A.append([0, 0, 0, x, y, 1, -x * yp, -y * yp, -yp])

    A = np.asarray(A, dtype=float)
    _, _, vh = np.linalg.svd(A)
    H = vh[-1].reshape(3, 3)
    H = H / H[2, 2]
    #############################################################################
    #                             END OF YOUR CODE                              #
    #############################################################################

    return H
```

### (a) Implement the following function to compute SIFT key points and features for a given image. Run the plotting code to visualize the key points and include them in the notebook.

Hint: you will need `cv2.SIFT_create` and some related functions

```python
def get_sift_features(img):
    """
    Compute SIFT features using cv2 library functions.
    Use default parameters when computing the keypoints.

    Input:
        img: cv2 image
    Returns:
        keypoints: a list of cv2 keypoints
        descriptors: a list of SIFT descriptors
    """

    keypoints = None
    descriptors = None

    #############################################################################
    cache = getattr(get_sift_features, '_cache', {})
    cache_key = id(img)
    if cache_key in cache:
        return cache[cache_key]

    img_for_sift = img
    if img_for_sift.dtype != np.uint8:
        scale = 255.0 if img_for_sift.max() <= 1.0 else 1.0
        img_for_sift = np.clip(img_for_sift * scale, 0, 255).astype(np.uint8)
    if img_for_sift.ndim == 3:
        gray = cv2.cvtColor(img_for_sift, cv2.COLOR_RGB2GRAY)
    else:
        gray = img_for_sift

    sift = cv2.SIFT_create()
    keypoints, descriptors = sift.detectAndCompute(gray, None)

    cache[cache_key] = (keypoints, descriptors)
    get_sift_features._cache = cache
    #############################################################################
    #                             END OF YOUR CODE                              #
    #############################################################################

    return keypoints, descriptors
```

### (b) Implement the following function to compute keypoint correspondences between the feature descriptors from two source images using the Lowe's ratio test. Run the plotting code to visualize the resulting correspondences and include them in the notebook.

Hint: you may need to use `cv2.DescriptorMatcher_create`, `knnMatch`, and some related functions

```python
def match_keypoints(desc_1, desc_2, ratio=0.75):
    """
    Compute matches between feature descriptors of two images using Lowe's ratio test.

    Input:
        desc_1, desc_2: list of feature descriptors
        ratio (optional): Lowe's ratio test threshold
    Return:
        matches: list of feature matches
    """

    matches = None

    #############################################################################
    if desc_1 is None or desc_2 is None:
        return []

    matcher = cv2.DescriptorMatcher_create(cv2.DescriptorMatcher_BRUTEFORCE)
    knn_matches = matcher.knnMatch(desc_1, desc_2, k=2)

    matches = []
    for pair in knn_matches:
        if len(pair) < 2:
            continue
        m, n = pair
        if m.distance < ratio * n.distance:
            matches.append(m)

    matches = sorted(matches, key=lambda m: m.distance)
    #############################################################################
    #                             END OF YOUR CODE                              #
    #############################################################################

    return matches
```

### (c) Implement the following function that fits a homography using RANSAC, which gives you better robustness to outliers than `fit_homography(pts1, pts2)`. You are responsible for finding the optimal hyperparameters for RANSAC.

Hint You can call `fit_homography(pts1, pts2)` inside the inner loop of RANSAC. Use the hyperparameters specified in the pdf file.

```python
def transform_ransac(x1, x2, verbose=False):
    """
    Implements RANSAC to estimate homography matrix.

    Input:
        x1, x2: (N, 2) matrices
    Return:
        best_model: homography matrix with most inliers
    """

    best_model = None

    #############################################################################
    if x1.shape != x2.shape:
        raise ValueError('x1 and x2 must have the same shape')
    if x1.shape[0] < 4:
        raise ValueError('At least 4 point correspondences are required')

    sample_size = 4
    num_iters = 200
    inlier_threshold = 4.0
    rng = np.random.default_rng(442)
    x1_h = to_homog(x1)
    best_inliers = None
    best_inlier_count = 0
    best_error = np.inf

    def reprojection_errors(H):
        projected_h = (H @ x1_h.T).T
        errors = np.full(x1.shape[0], np.inf, dtype=float)
        valid = np.abs(projected_h[:, 2]) > 1e-8
        if np.any(valid):
            projected = projected_h[valid, :2] / projected_h[valid, 2:3]
            errors[valid] = np.linalg.norm(projected - x2[valid], axis=1)
        return errors

    for _ in range(num_iters):
        idx = rng.choice(x1.shape[0], size=sample_size, replace=False)
        H_candidate = fit_homography(x1[idx], x2[idx])
        errors = reprojection_errors(H_candidate)
        inliers = errors < inlier_threshold
        if np.count_nonzero(inliers) < 4:
            continue

        H_candidate = fit_homography(x1[inliers], x2[inliers])
        errors = reprojection_errors(H_candidate)
        inliers = errors < inlier_threshold
        inlier_count = np.count_nonzero(inliers)
        if inlier_count < 4:
            continue

        mean_error = errors[inliers].mean()
        if inlier_count > best_inlier_count or (inlier_count == best_inlier_count and mean_error < best_error):
            best_model = H_candidate
            best_inliers = inliers
            best_inlier_count = inlier_count
            best_error = mean_error

    if best_model is None:
        best_model = fit_homography(x1, x2)
    elif best_inliers is not None and np.count_nonzero(best_inliers) >= 4:
        best_model = fit_homography(x1[best_inliers], x2[best_inliers])

    if verbose:
        print(f'RANSAC inliers: {best_inlier_count}/{x1.shape[0]}')
    #############################################################################
    #                             END OF YOUR CODE                              #
    #############################################################################

    return best_model
```

### (d) Implement the following function that fits a homography given two input images using your functions from the previous parts.

```python
def homography_from_image_pairs(img1, img2):
    """
    Given a pair of overlapping images, generate the homography matrix H.

    Input:
        img1, img2: cv2 images
    Return:
        H: computed homography matrix
    """

    H = None

    #############################################################################
    cache = getattr(homography_from_image_pairs, '_cache', {})
    cache_key = (id(img1), id(img2))
    if cache_key in cache:
        return cache[cache_key].copy()

    kp_1, desc_1 = get_sift_features(img1)
    kp_2, desc_2 = get_sift_features(img2)
    matches = match_keypoints(desc_1, desc_2)
    if len(matches) < 4:
        raise ValueError('Not enough matches to estimate a homography')

    inlier_threshold = 4.0
    matches_ransac = matches[:150]
    pts_1 = np.float32([kp_1[m.queryIdx].pt for m in matches_ransac])
    pts_2 = np.float32([kp_2[m.trainIdx].pt for m in matches_ransac])
    H = transform_ransac(pts_2, pts_1)

    pts_1_all = np.float32([kp_1[m.queryIdx].pt for m in matches])
    pts_2_all = np.float32([kp_2[m.trainIdx].pt for m in matches])
    projected_h = (H @ to_homog(pts_2_all).T).T
    errors = np.full(pts_2_all.shape[0], np.inf, dtype=float)
    valid = np.abs(projected_h[:, 2]) > 1e-8
    if np.any(valid):
        projected = projected_h[valid, :2] / projected_h[valid, 2:3]
        errors[valid] = np.linalg.norm(projected - pts_1_all[valid], axis=1)
    inliers = errors < inlier_threshold
    if np.count_nonzero(inliers) >= 4:
        H = fit_homography(pts_2_all[inliers], pts_1_all[inliers])

    cache[cache_key] = H.copy()
    homography_from_image_pairs._cache = cache
    #############################################################################
    #                             END OF YOUR CODE                              #
    #############################################################################

    return H
```

### (e) Implement the following function that produces a panorama from a list of overlapping images using your functions from the previous parts. Run the plotting code to visualize the resulting panorama and include them in the notebook.

Hints: You may use `cv2.warpPerspective`.



```python
def panoramic_stitching(*imgs):
    """
    Given a list of overlapping images, generate a panoramic image.

    Input:
        imgs: cv2 images
    Return:
        final_img: cv2 image of panorama
    """

    final_img = None

    #############################################################################
    if len(imgs) == 0:
        raise ValueError('At least one image is required')
    if len(imgs) == 1:
        return imgs[0].copy()

    pairwise_H = [homography_from_image_pairs(imgs[i], imgs[i + 1]) for i in range(len(imgs) - 1)]

    homographies = [np.eye(3)]
    for H_pair in pairwise_H:
        homographies.append(homographies[-1] @ H_pair)

    all_corners = []
    for img, H in zip(imgs, homographies):
        h, w = img.shape[:2]
        corners = np.array([[0, 0], [w - 1, 0], [w - 1, h - 1], [0, h - 1]], dtype=float)
        warped_corners = to_cart((H @ to_homog(corners).T).T)
        all_corners.append(warped_corners)

    all_corners = np.vstack(all_corners)
    min_x = int(np.floor(all_corners[:, 0].min()))
    min_y = int(np.floor(all_corners[:, 1].min()))
    max_x = int(np.ceil(all_corners[:, 0].max()))
    max_y = int(np.ceil(all_corners[:, 1].max()))

    translate = np.array([
        [1.0, 0.0, -min_x],
        [0.0, 1.0, -min_y],
        [0.0, 0.0, 1.0],
    ])

    width = max_x - min_x + 1
    height = max_y - min_y + 1

    if imgs[0].ndim == 2:
        accumulator = np.zeros((height, width), dtype=np.float32)
        weights = np.zeros((height, width), dtype=np.float32)
    else:
        accumulator = np.zeros((height, width, imgs[0].shape[2]), dtype=np.float32)
        weights = np.zeros((height, width, 1), dtype=np.float32)

    for img, H in zip(imgs, homographies):
        warp_H = translate @ H
        warped = cv2.warpPerspective(img.astype(np.float32), warp_H, (width, height))
        mask = cv2.warpPerspective(np.ones(img.shape[:2], dtype=np.uint8), warp_H, (width, height))

        if imgs[0].ndim == 2:
            mask = (mask > 0).astype(np.float32)
            accumulator += warped * mask
            weights += mask
        else:
            mask = (mask > 0).astype(np.float32)[..., None]
            accumulator += warped * mask
            weights += mask

    final_img = accumulator / np.maximum(weights, 1.0)
    if np.issubdtype(imgs[0].dtype, np.integer):
        final_img = np.clip(final_img, 0, 255).astype(imgs[0].dtype)
    else:
        final_img = final_img.astype(imgs[0].dtype)

    occupied = weights if imgs[0].ndim == 2 else weights[..., 0]
    ys, xs = np.where(occupied > 0)
    if ys.size > 0 and xs.size > 0:
        final_img = final_img[ys.min():ys.max() + 1, xs.min():xs.max() + 1]
    #############################################################################
    #                             END OF YOUR CODE                              #
    #############################################################################

    return final_img
```
