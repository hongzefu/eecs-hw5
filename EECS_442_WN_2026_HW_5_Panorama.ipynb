# Problem 1: Homography Estimation

First, we will implement a function that computes a homography matrix $H$ from a set of corresponding points between two images. We'll estimate $H$ using two methods: 1) an analytic (linear) method and 2) a nonlinear least squares optimization method. For a pair of corresponding points, $p_i$ and $p_i'$, we solve for a homography matrix $H$ such that $p_i' = H p_i$:

$$
\left[\begin{array}{c}{x'} \\ {y'} \\ {1}\end{array}\right] = \left[\begin{array}{lll}{a} & {b} & {c} \\ {d} & {e} & {f} \\ {g} & {h} & {i}\end{array}\right] \cdot\left[\begin{array}{l}{x} \\ {y} \\ {1}\end{array}\right]
$$


We will work with two sets of 2D points representing the shape of the University of Michigan’s "M" logo, but at different orientations (see below for visualization). One set serves as the original points, while the other represents the destination points. Our goal is to compute the homography that maps the original points to the destination points, enabling us to warp the original shape accordingly.
